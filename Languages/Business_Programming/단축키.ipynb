자주 쓰는 편집용 단축키 (Help - Keyboard Shortcuts)
=================

1) 셀 선택 모드 (command mode, 파란색 창)
====================================

enter: 입력 셀로 전환

a: 위에 새 창 만들기

b: 아래에 새 창 만들기

p: 선택 셀 아래에 붙여넣기

c : copy selected cell

shift-v : paste cell above

v : paste cell below

shift+m: 선택 셀 아래 셀과 합치기

x: 셀 잘라내기

dd: 셀 지우기

o: 실행결과 열기/닫기

s: 파일 저장

m: 마크다운(주석)으로 변경

y: 코드로 변경

1>2>3>4: 텍스트 크기 순서

<br> : 줄바꾸기

- , 1. : 글머리 넣기

--- : 수평선 그어주기 (단락 구분)





2) 셀 편집 모드 (Edit mode, 초록색 창)
================================

shift + enter: 코드 실행, running, execution

선택 셀의 코드 전체 선택
[ctrl] + [a]

선택 셀 내 실행 취소
[ctrl] + [z]

선택 셀 내 다시 실행
[ctrl] + [y]

커서 위치 라인 주석처리
[ctrl] + [/]

코드 내 마크다운(주석): #

선택 셀 코드 실행
[ctrl] + [enter]

선택 셀 코드 실행 후 다음 Cell로 이동 (없으면 새로 추가)
[shift] + [enter]

커서 위치에서 셀 둘로 나누기
[shift] + [ctrl] + [-]

파일 저장
[ctrl] + [s]

셀 선택 모드로 돌아가기
[esc] 또는 [ctrl] + [m]
