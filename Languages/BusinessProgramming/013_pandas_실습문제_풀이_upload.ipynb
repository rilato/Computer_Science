```python
#%% DF24. 다음의 값으로 구성된 DataFrame을 생성하시오 

"""Original DataFrame
col1 col2 col3
0 1 4 7
1 4 5 8
2 3 6 9
3 4 7 0
4 5 8 1
Rows for colum1 value == 4
"""
import pandas as pd
import numpy as np
df_dic={'col1': [1, 4, 3, 4, 5], 'col2': [4, 5, 6, 7, 8], 'col3': [7, 8, 9, 0, 1]}
df = pd.DataFrame(df_dic)
print(df)
```

```python
#%% baseball.csv 로부터 데이터를 읽어와 dataframe으로 저장하고 컬럼 labels,컬럼 labels 수, 데이터수, 로우와 컬럼수를 출력하시오.


import pandas as pd
import numpy as np

df = pd.read_csv("baseball.csv")
print(df.columns)
print(len(df.columns))
print(df.size)
print(df.shape)
```

```python
#%% DF36. baseball 데이터 df에서 256번째, 478번째 row를 없앤 후 다시 출력하시오 

import pandas as pd
import numpy as np

df = df.drop([256,478], axis=0)
display(df)
```

```python
#%% baseball 데이터 df에서 선수들의 키의 평균은 얼마인가? 

a = sum(df.Height)/len(df.Height)
avg = df.Height.mean()

print(a)
print(avg)
```

```python
#%% baseball 데이터 df에서 몸무게가 190이상이고 200 이하인 선수는 몇명인가?

a = len(df[(df.Weight>=190) & (df.Weight<=200)])

print(a)
```

```python
#%% baseball 데이터 df에서 20대 이상, 30대 이상 선수는 각각 몇명인가?

a = len(df[df.Age>=20])
b = len(df[df.Age>=30])

print(a)
print(b)
```
